```python
import pandas as pd

print("Cargando base de datos de texto")
df_textos = pd.read_csv("all_tickets_processed_improved_v3.csv")


df_textos = df_textos.dropna()

print(f"\nTotal de tickets de texto listo: {len(df_textos)}")
print("=== TOP 5 CATEGORÍAS MÁS COMUNES ===")
print(df_textos['Topic_group'].value_counts().head(5))

print("=== Ejemplo de un ticket ===")
print(df_textos['Document'].iloc[0])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

X_texto = df_textos['Document']
y_categoria = df_textos['Topic_group']

print("Convirtiendo todo el texto a una matriz matematica (TF-IDF)...")

vectorizador = TfidfVectorizer(max_features=5000)


X_numeros = vectorizador.fit_transform(X_texto.astype(str))

X_train, X_test, y_train, y_test = train_test_split(X_numeros, y_categoria, test_size=0.20, random_state=42)

print("\n¡Transformación matemática completada con éxito!")
print(f"Dimensiones de X_train (Material de estudio): {X_train.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

print("Creando el modelo (Regresión Logística)...")

modelo_nlp = LogisticRegression(max_iter=1000, random_state=42)

print("Entrenando la IA para que lea los 38,269 tickets... (Esto tomara unos segundos)")

modelo_nlp.fit(X_train, y_train)
print("Entrenamiento completado")

print("\nResolviendo el examen final con tickets que nunca ha visto...")

predicciones_texto = modelo_nlp.predict(X_test)


calificacion_nlp = accuracy_score(y_test, predicciones_texto) * 100

print(f"\n=== RESULTADOS DEL CLASIFICADOR DE TEXTO ===")
print(f"Precision (Accuracy): {calificacion_nlp:.2f}%")


print("\nReporte detallado por Categoria (Departamento):")
print(classification_report(y_test, predicciones_texto))
```

```python

mi_queja = ["I forgot my password and cannot login."]


queja_traducida = vectorizador.transform(mi_queja)

departamento_asignado = modelo_nlp.predict(queja_traducida)

print("=== SISTEMA DE ENRUTAMIENTO AUTOMATICO ===")
print(f"Mensaje del usuario: '{mi_queja[0]}'")
print(f"Asignado automáticamente al departamento: {departamento_asignado[0]}")
```
